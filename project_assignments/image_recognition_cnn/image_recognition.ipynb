# Neural Networks Project: Assignment 2 
Image Recognition with Convolutional Neural Networks (CNNs)

Images resized to 224x224 as pretrained models like ResNet and VGG typically expect this input size. Images are converted to RGB to use all three color channels. 

```python
import os
from PIL import Image
import random
import shutil

RAW_DIR = "data/raw"
PROCESSED_DIR = "data/processed"
IMG_SIZE = (224, 224)

def preprocess_all():
    for class_name in os.listdir(RAW_DIR):
        raw_class_path = os.path.join(RAW_DIR, class_name)
        processed_class_path = os.path.join(PROCESSED_DIR, class_name)

        os.makedirs(processed_class_path, exist_ok=True)

        for img_name in os.listdir(raw_class_path):
            raw_path = os.path.join(raw_class_path, img_name)
            processed_path = os.path.join(processed_class_path, img_name)

            if os.path.exists(processed_path):
                continue

            try:
                img = Image.open(raw_path).convert("RGB")
                img = img.resize(IMG_SIZE)
                img.save(processed_path)

            except Exception as e:
                print(f"Error with {img_name}: {e}")
```

```python
preprocess_all()
```

```python
source_dir = PROCESSED_DIR 
dest_dir = "data/split"
for class_name in os.listdir(source_dir):
        images = os.listdir(os.path.join(source_dir, class_name))
        random.shuffle(images)

        n = len(images)
        train = images[:int(0.7*n)]
        val   = images[int(0.7*n):int(0.85*n)]
        test  = images[int(0.85*n):]

        for split, split_data in zip(['train','val','test'], [train, val, test]):
            for img in split_data:
                src = os.path.join(source_dir, class_name, img)
                dst = os.path.join(dest_dir, split, class_name, img)
                os.makedirs(os.path.dirname(dst), exist_ok=True)
                shutil.copy(src, dst)
```
